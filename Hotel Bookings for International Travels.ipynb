```python
# Function to plot distributions
def plot_distributions(df, column):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title('Distribution of ' + column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Plot distribution of bookings across months
if 'Booking Month' in data.columns:
    plot_distributions(data, 'Booking Month')
else:
    print('Month column not found, please provide the correct column name for booking months.')
```

```python
# Convert 'Date of Booking' to datetime and extract the month

data['Booking Month'] = pd.to_datetime(data['Date of Booking']).dt.month

import calendar

# Convert 'Booking Month' numeric values to month names
data['Booking Month'] = data['Booking Month'].apply(lambda x: calendar.month_name[x])
```

```python
# Plot the distribution of bookings across months
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=data, x='Booking Month', palette="viridis")

# Add labels to the top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Distribution of Bookings Across Months')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
# Group the data by 'Booking Month' and count the number of bookings for each month
monthly_bookings = data.groupby('Booking Month').size()

# Identify peak booking periods (months with the highest number of bookings)
peak_months = monthly_bookings.sort_values(ascending=False).head()

# Identify low booking periods (months with the lowest number of bookings)
low_months = monthly_bookings.sort_values(ascending=True).head()

# Print the peak and low booking periods
print('Peak booking months (Month: Bookings):')
print(peak_months)
print('\
Low booking months (Month: Bookings):')
print(low_months)

# Check for specific destinations or origin countries that drive seasonal variations

# Group by 'Booking Month' and 'Origin Country' and count the number of bookings
origin_monthly = data.groupby(['Booking Month', 'Origin Country']).size().reset_index(name='Bookings')
# Group by 'Booking Month' and 'Destination Country' and count the number of bookings
destination_monthly = data.groupby(['Booking Month', 'Destination Country']).size().reset_index(name='Bookings')
# Find the top 5 origin countries for peak months
top_origin_for_peak = origin_monthly[origin_monthly['Booking Month'].isin(peak_months.index)].sort_values(by='Bookings', ascending=False).head()
# Find the top 5 destination countries for peak months
top_destination_for_peak = destination_monthly[destination_monthly['Booking Month'].isin(peak_months.index)].sort_values(by='Bookings', ascending=False).head()
print('\
Top 5 origin countries during peak months:')
print(top_origin_for_peak)
print('\
Top 5 destination countries during peak months:')
print(top_destination_for_peak)
```

```python
from scipy.stats import chi2_contingency

# Create age groups
bins = [0, 18, 30, 40, 50, 60]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Analyze the correlation between demographics and room types
room_type_crosstab = pd.crosstab(index=data['Age Group'], columns=data['Rooms'])
chi2, p, dof, ex = chi2_contingency(room_type_crosstab)

# Analyze the correlation between demographics and destination countries
destination_country_crosstab = pd.crosstab(index=data['Age Group'], columns=data['Destination Country'])
chi2_dest, p_dest, dof_dest, ex_dest = chi2_contingency(destination_country_crosstab)

# Visualize the crosstab of age group and room types
plt.figure(figsize=(12, 8))
sns.heatmap(room_type_crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Age Group vs. Room Types')
plt.xlabel('Room Types')
plt.ylabel('Age Group')
plt.show()

# Visualize the crosstab of age group and destination countries
plt.figure(figsize=(12, 8))
sns.heatmap(destination_country_crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Age Group vs. Destination Countries')
plt.xlabel('Destination Countries')
plt.ylabel('Age Group')
plt.show()

# Output the results of the chi-squared tests
print('Chi-squared test for Age Group vs. Room Types:\
Chi2:', chi2, '\
P-value:', p, '\
Degrees of Freedom:', dof)
print('\
Chi-squared test for Age Group vs. Destination Countries:\
Chi2:', chi2_dest, '\
P-value:', p_dest, '\
Degrees of Freedom:', dof_dest)
```

The chi-squared tests have been conducted to analyze the correlation between demographics and room types, as well as demographics and destination countries. Here are the results:

For Age Group vs. Room Types:

Chi-squared value: 7.0549
P-value: 0.6314
Degrees of Freedom: 9
For Age Group vs. Destination Countries:

Chi-squared value: 89.1275
P-value: 0.0042
Degrees of Freedom: 57
    
    
The P-value for the correlation between age group and room types is high (0.6314), indicating that there is no significant correlation between these variables. In contrast, the P-value for the correlation between age group and destination countries is low (0.0042), suggesting a significant correlation.

The heatmaps show the distribution of bookings across different room types and destination countries for each age group. The significant correlation between age groups and destination countries suggests that certain age groups may prefer certain destinations.
